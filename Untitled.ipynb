```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=pd.read_csv("./Training Data/Diabetes_XTrain.csv")
Y=pd.read_csv("./Training Data/Diabetes_YTrain.csv")
test_data=pd.read_csv("./Test Cases/Diabetes_Xtest.csv")
```

```python
print(X.shape)
print(Y.shape)
print(test_data.shape)
```

```python
print(type(Y))
Y=np.reshape(Y,(-1,))
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**.5

def knn(X,Y,test_point,k=17):
    
    d = []
    m = X.shape[0]
    for i in range(m):
        dis = dist(test_point,X[i])
        d.append((dis,Y[i]))
    
    d.sort()
    d = np.array(d[:k])[:,1]
    uniq,freq = np.unique(d,return_counts=True)
    idx = np.argmax(freq)
    label = uniq[idx]
    
    return int(label)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
outputs=[]
m=test_data.shape[0]
for i in range(m):
    outputs.append(knn(X, Y, test_data[i]))
```

```python
outputs=np.array(outputs)
outputs.shape
```

```python
submit=pd.DataFrame(outputs, columns=["Outcome"])
```
